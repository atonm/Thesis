```python
n1 = len(A)
n2 = len(B)

#平均値
mu = (n1 * n2)/2
print('平均値：', mu)

#標準偏差
sigma = np.sqrt(n1 * n2 * (n1 + n2 + 1)/12)

U=result.statistic
print('標準偏差：', round(sigma,2))
z = abs((U - mu)/sigma)
print('z:',round(z,3))
```
