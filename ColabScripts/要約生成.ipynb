```python
from langchain import OpenAI, PromptTemplate, LLMChain
from langchain.chat_models import ChatOpenAI
from langchain.chains.mapreduce import MapReduceChain
from langchain.prompts import PromptTemplate, load_prompt
from langchain.text_splitter import CharacterTextSplitter
from langchain.docstore.document import Document
from langchain.chains.summarize import load_summarize_chain
#langchain.verbose = False
#from langchain.chains import ReduceDocumentsChain, MapReduceDocumentsChain

# LLMの準備
chat_llm = ChatOpenAI(
    model_name="gpt-4-1106-preview",
    temperature=0)

# 長文を分割するオブジェクトを定義
text_splitter = CharacterTextSplitter()

#要約するファイルを開く
with open("/content/rur_edited2.txt") as f:
    state_of_the_union = f.read()

text_splitter = CharacterTextSplitter(
    separator = "\n",  # セパレータ
    chunk_size = 3000,  # チャンクの文字数
    chunk_overlap = 0,  # チャンクオーバーラップの文字数
)
# 要約したい長文の分割
texts = text_splitter.split_text(state_of_the_union)
print(texts)


# 分割した長文を各文章ごとにDocumentオブジェクト化
docs = [Document(page_content=t) for t in texts]
print(docs)
```

```python
chain = load_summarize_chain(
    llm=chat_llm,
    chain_type="map_reduce",
    return_intermediate_steps=True, # 分割された文章ごとの要約結果を返すか
    map_prompt=load_prompt("map_prompt.json"), # load_promptはpromtの設定をファイルで読み込んでいる。json, py, yamlファイルで読み込める
    combine_prompt=load_prompt("combine_prompt.json"))
result=chain({"input_documents": docs, "token_max":4097}, return_only_outputs=True)
print(result["output_text"])
#print(chain.run(docs))
```

```python
from langchain.chains.summarize import load_summarize_chain
prompt="あなたの仕事は文章を全て読んで要約することです。台本を与えるので、そのストーリーを要約してください。"
prompt2="与えられたものをそのまま出力してください。"
chain = load_summarize_chain(llm=chat_llm,
														 chain_type="map_reduce", # 要約の仕方 stuff, map_reduce, refineから選ぶ
														 return_intermediate_steps=True, # 分割された文章ごとの要約結果を返すか
                             map_prompt=prompt,
                             combine_prompt=prompt2)


# 要約の実行
print(chain.run(docs))
#print(chain.predict(input=messages))
```
